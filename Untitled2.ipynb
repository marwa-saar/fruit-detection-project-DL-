```python
from ultralytics import YOLO
from pathlib import Path
```

```python
# تحميل موديل جاهز مدرّب على dataset الفواكه
model = YOLO("weights/yolov8l-fruits.pt")
```

```python
# مسار Dataset
dataset_path = Path(r"C:/Users/USER PC/.cache/kagglehub/datasets/lakshaytyagi01/fruit-detection")
data_yaml_path = dataset_path / "data.yaml"

# التحقق أن الملفات موجودة
if not dataset_path.exists():
    raise FileNotFoundError(f"❌ Dataset not found at {dataset_path}")
if not data_yaml_path.exists():
    raise FileNotFoundError(f"❌ data.yaml not found at {data_yaml_path}")

# مسار الأوزان المدربة مسبقًا
pretrained_weights = Path("weights/yolov8l-fruits.pt")
if not pretrained_weights.exists():
    raise FileNotFoundError(f"❌ Pretrained weights not found at {pretrained_weights}")
```

```python
# تحميل موديل جاهز مدرّب على dataset الفواكه
model = YOLO(pretrained_weights)
```

```python
# تدريب إضافي على بياناتك
model.train(
    data=str(data_yaml_path),  # مسار ملف YAML للبيانات
    epochs=10,                 # عدد مرات مرور البيانات
    imgsz=512,                 # حجم الصور أثناء التدريب
    batch=8,                   # عدد الصور في كل batch
    name="fruit_detection_finetune"  # اسم المشروع/المجلد الناتج
)
```
