```python
# ===============================
# 2- استيراد المكتبة
# ===============================
from ultralytics import YOLO
import os
```

```python
# ===============================
# 3- تحديد مسار الديتا سيت
# ===============================
dataset_path = r"C:\Users\USER PC\.cache\kagglehub\datasets\lakshaytyagi01\fruit-detection"
data_yaml_path = os.path.join(dataset_path, "data.yaml")
```

```python
# ===============================
# 4- التحقق من الملفات (اختياري)
# ===============================
print("محتويات مجلد الديتا سيت:", os.listdir(dataset_path))
print("أول 5 صور في train:", os.listdir(os.path.join(dataset_path, "train"))[:5])
```

```python
# ===============================
# 5- تحميل موديل YOLOv8s الجاهز
# ===============================
model = YOLO(os.path.join(dataset_path, 'yolov8s.pt'))
```

```python
# ===============================
# 6- تدريب الموديل مع Early Stopping
# ===============================
# ملاحظة: النسخة الحديثة تستخدم 'patience' مباشرة لتفعيل الإيقاف المبكر
results = model.train(
    data=data_yaml_path,
    epochs=50,      # أكبر عدد من epochs
    imgsz=416,       # حجم الصور
    batch=4,        # حجم الباتش
    patience=10      # الإيقاف المبكر بعد 10 epochs بدون تحسن
)
```

```python
# ===============================
# 7- الحصول على أفضل موديل
# ===============================
best_model_path = results.best
print("أفضل موديل محفوظ في:", best_model_path)
```

```python
# ===============================
# 8- تحميل أفضل موديل بعد التدريب
# ===============================
best_model = YOLO(best_model_path)
```

```python
# ===============================
# 9- اختبار الموديل على صورة من مجلد test
# ===============================
test_img_path = os.path.join(dataset_path, "test", "images", os.listdir(os.path.join(dataset_path, "test", "images"))[0])
results = best_model.predict(source=test_img_path, conf=0.5)
results.show()
```

```python

# ===============================
# 10- تصدير الموديل بصيغ أخرى (اختياري)
# ===============================
best_model.export(format='pt', name='fruit_detector_final.pt')     # حفظ بصيغة PyTorch
best_model.export(format='onnx', name='fruit_detector_final.onnx') # حفظ بصيغة ONNX
```
